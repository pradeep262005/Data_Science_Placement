```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
```

```python
df['Performance in a entrance']=encoder.fit_transform(df['Performance in a entrance'])
df['Performance in a entrance'].unique()
```

```python
data=df.drop(['Parental Income','Distance from college','Internet availability at home(Yes/No)','Final Year result'],axis=1)
input=data
print("The Input Data are")
print(input.columns)
```

```python
output=df['Final Year result']
print("The Output Data is")
print(output.name)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.2,random_state=5)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
input_data = pd.DataFrame([[55, 4, 18, 0]], columns=['High School GPA','No.of Extracurricular activites',
       'No.of hours studied per week','Performance in a entrance'])
input_data2 = pd.DataFrame([[46, 4, 18, 0]], columns=['High School GPA','No.of Extracurricular activites',
       'No.of hours studied per week','Performance in a entrance'])
prediction = model.predict(input_data)
prediction2=model.predict(input_data2)
print("The result according to the given data is", prediction)
print("The result according to the given data is", prediction2)
```
